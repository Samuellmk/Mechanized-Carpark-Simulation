```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the mean (scale parameter) of the Weibull distribution
mean_hours = 4.5

# Calculate the shape parameter (k) using a desired shape parameter or by estimating it
shape_parameter = 1.5

# Create a Weibull distribution object
weibull_dist = stats.weibull_min(shape_parameter, scale=mean_hours)

# Generate random samples from the Weibull distribution
num_samples = 1000
samples = weibull_dist.rvs(size=num_samples)

# Calculate the PDF values for a range of x values
x = np.linspace(0, 10, 1000)
pdf_values = weibull_dist.pdf(x)

# Create a histogram of the generated samples
plt.hist(samples, bins=13, density=True, alpha=0.5, label='Histogram of Samples')

# Plot the PDF of the Weibull distribution
plt.plot(x, pdf_values, 'r-', lw=2, label='PDF of Weibull Distribution')

plt.xlabel('Hours')
plt.ylabel('Probability Density')
plt.title('Weibull Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
